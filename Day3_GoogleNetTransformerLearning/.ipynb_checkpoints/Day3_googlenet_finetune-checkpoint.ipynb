```python
import glob
import os.path
import random
import numpy as np
import tensorflow as tf
from tensorflow.python.platform import gfile
```

```python
'''
数据超参数
'''
MODEL_DIR = './model/'
MODEL_FILE = 'tensorflow_inception_graph.pb'
CACHE_DIR = './data/tmp/bottleneck/'
INPUT_DATA = './data/flower_photos/'
VALIDATION_PERCENTAGE = 10
TEST_PERCENTAGE = 10
```

```python
'''
预加载模型参数
'''
BOTTLENECK_TENSOR_SIZE = 2048 # 这个是什么意思？
BOTTLENECK_TENSOR_NAME = 'pool_3/_reshape:0' # 瓶颈层输出的张量
JPEG_DATA_TENSOR_NAME = 'DecodeJpeg/contents:0' # 输入张量名字
```

```python
'''
训练参数
'''
LEARNING_RATE = 0.01
STEPS = 1000
BATCH = 100
CHECKPOINT_EVERY = 100
NUM_CHECKPOINTS =5
```

```python
'''
读取数据，并按照训练，验证，测试集分开
'''
def create_image_lists(validation_percentage,test_percentage):
    result = {}
    sub_dirs = [x[0] for x in os.walk(INPUT_DATA)]# 获取所有子目录
    is_root_dir = True # 第一个为当前目录，需要忽略
    
    # 对每个子目录进行操作
    for sub_dir in sub_dirs:
        if is_root_dir:
            is_root_dir=False
            continue
        
        # 获取当前子目录下所有图片
        extensions = {'jpg','jpeg','JPG','JPEG'}
        file_list = []
        dir_name = os.path.basename(sub_dir)
        for extension in extensions:
            file_glob = os.path.join(INPUT_DATA,dir_name,'*.'+extension)
            # 这里注意extend和append的区别
            file_list.extend(glob.glob(file_glob))
        if not file_list:
            continue
            
        # 将当前类别的图片随机分为训练数据集，测试数据集，验证数据集
        label_name = dir_name.lower()# 通过dir名字获取当前图片的label
        training_images = []
        testing_images = []
        validation_images = []
        for file_name in file_list:
            base_name = os.path.basename(file_name) # 获取图片名称
            chance = np.random.randint(100)
            if chance < validation_percentage:
                validation_images.append(base_name)
            elif chance < (validation_percentage+test_percentage):
                testing_images.append(base_name)
            else:
                training_images.append(base_name)
        result[label_name]={
            'dir':dir_name,
            'training':training_images,
            'testing':testing_images,
            'validation':validation_images
        }
        
    return result
```

```python
'''
通过：
    类别名称
    所属数据集
    图片编号
来获得一张图片的地址
'''
def get_image_path(image_lists,image_dir,label_name,index,category):
    # 获得给定类别中的所有图片
    label_lists = image_lists[label_name]
    
    # 根据所属数据集的名称获得该集合中的全部图片
    category_list = label_lists[category]
    
    # 规范索引
    mod_index = index % len(category_list)
    
    # 获得图片文件名
    base_name = category_list[mod_index]
    
    # 获得当前类别目录名
    sub_dir = label_lists['dir']
    
    # 图片绝对路径
    full_path = os.path.join(image_dir,sub_dir,base_name)
    return full_path
```

```python
'''
通过：
    类别名称
    所属数据集
    图片编号
获得特征向量值的地址
'''
def get_bottleneck_path(image_lists,label_name,index,category):
    return get_image_path(image_lists,CACHE_DIR,
                          label_name,index,category)+'.txt'
```

```python
'''
使用inception——v3处理图片， 获得特征向量
'''
def run_bottleneck_on_image(sess,image_data,image_data_tensor,
                            bottleneck_tensor):
    bottleneck_values = sess.run(bottleneck_tensor,
                                 feed_dict={image_data_tensor:image_data})
    bottleneck_values = np.squeeze(bottleneck_values)
    return bottleneck_values
```

```python
'''
获取一张图片经过inception-v3模型处理后的特征向量
'''
def get_or_create_bottleneck(sess,image_lists,label_name,index,category,
                             jpeg_data_tensor,bottleneck_tensor):
    # 获取一张图片对应的特征向量文件路径
    label_lists = image_lists[label_name]
    sub_dir = label_lists['dir']
    sub_dir_path = os.path.join(CACHE_DIR,sub_dir)
    if not os.path.exists(sub_dir_path):
        os.makedirs(sub_dir_path)
    bottleneck_path = get_bottleneck_path(image_lists,label_name,index,category)
    
    # 如果特征向量文件不存在，则通过inception-v3模型进行计算并保存
    if not os.path.exists(bottleneck_path):
        # 获得图片原始路径
        image_path = get_image_path(image_lists,INPUT_DATA,label_name,index,category)
        # 获得图片内容（数据）
        image_data = gfile.FastGFile(image_path,'rb').read()
        # 通过inception-v3计算特征向量
        bottleneck_values = run_bottleneck_on_image(sess,image_data,
                                                    jpeg_data_tensor,bottleneck_tensor)
        bottleneck_string = ','.join(str(x) for x in bottleneck_values)
        with open(bottleneck_path,'w') as bottleneck_file:
            bottleneck_file.write(bottleneck_string)
    else:
        # 直接从文件中获得特征向量
        with open(bottleneck_path,'r') as bottleneck_file:
            bottleneck_string = bottleneck_file.read()
        bottleneck_values = [float(x) for x in bottleneck_string.split(',')]
    
    return bottleneck_values
```

```python
'''
随机获得一个batch的图片作为训练数据
'''
def get_random_cached_bottlenecks(sess,n_classes,image_lists,
                                  how_many,category,jpeg_data_tensor,
                                  bottleneck_tensor):
    bottlenecks = []
    ground_truths = []
    for _ in range(how_many):
        # 随机选择一个类别和图片编号加入当前的训练数据
        label_index = random.randrange(n_classes)
        label_name = list(image_lists.keys())[label_index]
        image_index = random.randrange(65535)
        bottleneck = get_or_create_bottleneck(sess,image_lists,
                                              label_name,image_index,
                                              category,jpeg_data_tensor,
                                              bottleneck_tensor)
        ground_truth = np.zeros(n_classes,dtype=np.float32)
        ground_truth[label_index]=1.0
        bottlenecks.append(bottleneck)
        ground_truths.append(ground_truth)
    return bottlenecks,ground_truths
```

```python
'''
获取全部的测试数据
'''
def get_test_bottlenecks(sess,image_lists,n_classes,jpeg_data_tensor,
                         bottleneck_tensor):
    bottlenecks = []
    ground_truths = []
    label_name_list = list(image_lists,keys())
    # 枚举每个类别和每个类别中的测试图片
    for label_index,label_name in enumerate(label_name_list):
        category = 'testing'
        for index,unused_base_name in enumerate(image_lists[label_name][category]):
            bottleneck = get_or_create_bottleneck(sess,image_lists,label_name,
                                                  index,category,
                                                  jpeg_data_tensor,
                                                  bottleneck_tensor)
            ground_truth = np.zeros(n_classes,dtype=np.float32)
            ground_truth[label_index] = 1.0
            bottlenecks.append(bottleneck)
            ground_truths.append(ground_truth)
    return bottlenecks,ground_truths
```

```python
def main():
    tf.reset_default_graph()
    # 读取所有图片
    image_lists = create_image_lists(VALIDATION_PERCENTAGE,TEST_PERCENTAGE)
    n_classes = len(image_lists.keys())
    
    with tf.Graph().as_default() as graph:
        # 读取训练好的inception-v3模型
        with gfile.FastGFile(os.path.join(MODEL_DIR,MODEL_FILE),'rb') as f:
            graph_def = tf.GraphDef()
            graph_def.ParseFromString(f.read())
            # 加载inception-v3模型，并返回数据的输入张量和瓶颈层输出张量
            bottleneck_tensor,jped_data_tensor = tf.import_graph_def(graph_def,return_elements=[BOTTLENECK_TENSOR_NAME,JPEG_DATA_TENSOR_NAME])
        
        # 定义神经网络的输入
        bottleneck_input = tf.placeholder(dtype=tf.float32,shape=[None,BOTTLENECK_TENSOR_SIZE],
                                          name='BottleneckInputPlaceholder')
        # 定义label的输入
        ground_truth_input = tf.placeholder(dtype=tf.float32,shape=[None,n_classes],
                                            name='GroundTruthInput')
        
        # 定义新的全连接层，来解决新的分类问题
        with tf.name_scope('final_training_ops'):
            weights =tf.Variable(tf.truncated_normal(shape=[BOTTLENECK_TENSOR_SIZE,n_classes],stddev=0.1))
            bias = tf.Variable(tf.constant(0.1,shape=[n_classes],dtype=tf.float32))
            '''
                从这里可以看出来，这套方法的思路是：
                    1、利用inception中全连接层以前的部分，得到每张图片长为2048的特征向量。
                    2、利用这些特征向量，训练一个新的一层，10分类的Dens层。
                    3、用的loss还是自己写的…
            '''
            logits = tf.matmul(bottleneck_input,weights)+bias
            final_tensor = tf.nn.softmax(logits)
        
        # 定义损失函数
        cross_entropy = tf.nn.softmax_cross_entropy_with_logits(logits=logits,
                                                                labels=ground_truth_input)
        cross_entropy_mean = tf.reduce_mean(cross_entropy)
        
```

```python
main()
```
